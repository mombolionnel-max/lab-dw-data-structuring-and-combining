```python
import pandas as pd
import numpy as np

url1= "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
df1=pd.read_csv(url1)
df1.rename({"ST":"State"}, axis=1, inplace=True)
df1.columns=df1.columns.str.lower()
df1
```

```python
url2= "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
df2=pd.read_csv(url2)
df2.rename({"ST":"State"}, axis=1, inplace=True)
df2.columns=df2.columns.str.lower()
df2
```

```python
url3= "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"
df3=pd.read_csv(url3)
df3.columns=df3.columns.str.lower()
df3
```

```python
new_df=pd.concat([df1, df2, df3], axis=0)
new_df
```

```python
new_df.dropna(axis=0 , how = "all")
```

```python
new_df["gender"].unique()
```

```python
new_df["state"].unique()
```

```python
def gender_change(value):
    if value == "Femal":
        return "F"
    elif value == "Male":
        return "M"        
    elif value == "female":
        return "F"
    elif value == "F":
        return "F"    
    elif value == "M":
        return "M"  
    else:
        return "nan"
    
new_df["gender"]=new_df["gender"].apply(gender_change)
```

```python
def translate_state(row):
    if row["state"] == "Cali":
        return "California"
    elif row["state"] =="WA":
        return "Washington"
    elif row["state"] =="AZ":
        return "Arizona"  
    else:
        return row["state"]
    
new_df["state"] = new_df.apply(translate_state, axis=1)
```

```python
def Bach(row):
    if row["education"] == "Bachelors":
        return "Bachelor" 
    else:
        return row["education"]
    
new_df["education"] = new_df.apply(Bach, axis=1)
```

```python
new_df["customer lifetime value"] = new_df["customer lifetime value"].str.replace('%', '')
new_df["customer lifetime value"] = pd.to_numeric(new_df["customer lifetime value"], errors='coerce')
```

```python
new_df["number of open complaints"] = new_df["number of open complaints"].str.split('/').str[1]
new_df["number of open complaints"] = pd.to_numeric(new_df["number of open complaints"], errors='coerce')
```
